# Worldwide Covid Vaccinations

```python
using System.Net.Http;
using System.Net.Http.Headers;
using System.Text.Json;
```

## Create an HttpClient for the network calls

```python
HttpClient client = new HttpClient();
client.DefaultRequestHeaders.Accept.Clear();
client.DefaultRequestHeaders.Accept.Add(new MediaTypeWithQualityHeaderValue("text/html"));
client.DefaultRequestHeaders.Accept.Add(new MediaTypeWithQualityHeaderValue("application/json"));
client.DefaultRequestHeaders.Add("User-Agent", "Notebooks (https://github.com/rprouse)");
```

## Get the daily code from the HTML page

```python
const string ROOT_URI = "https://static01.nyt.com/newsgraphics/2021/01/19/world-vaccinations-tracker/";

string html = await client.GetStringAsync("https://www.nytimes.com/interactive/2021/world/covid-vaccinations-tracker.html");
int start = html.IndexOf(ROOT_URI) + ROOT_URI.Length;
int end = html.IndexOf('/', start + 1);
string code = html.Substring(start, end - start);
display(code);
```

## Get the JSON Vaccination Data

```python
var json = await client.GetStringAsync($"{ROOT_URI}{code}/latest.json");
display(json);
```

## Let's Explore the Data

```python
#r "nuget:Microsoft.DotNet.Interactive.ExtensionLab,*-*"
```

```python
using System.Text.Json;

var vaccinations = JsonDocument.Parse(json);
display(vaccinations)
```

```python
vaccinations.ExploreWithNteract().Display();
```

```python
var country = vaccinations.RootElement.EnumerateArray()
  .FirstOrDefault(e => e.GetProperty("geoid").ToString() == "CAN");

display(country);
```

```python
var vaccinated = country.GetProperty("people_vaccinated").GetInt32();
var population = country.GetProperty("population").GetInt32();

var percent = (double)vaccinated/population*100d;

display(percent)
```
