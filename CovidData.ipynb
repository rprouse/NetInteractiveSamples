```python
using System.Net.Http;
using System.Net.Http.Headers;
using System.Text.Json;
```

## Create an HttpClient for the network calls

```python
HttpClient client = new HttpClient();
client.DefaultRequestHeaders.Accept.Clear();
client.DefaultRequestHeaders.Accept.Add(new MediaTypeWithQualityHeaderValue("text/html"));
client.DefaultRequestHeaders.Accept.Add(new MediaTypeWithQualityHeaderValue("application/json"));
client.DefaultRequestHeaders.Add("User-Agent", "Notebooks (https://github.com/rprouse)");
```

## Get the daily code from the HTML page

```python
const string ROOT_URI = "https://static01.nyt.com/newsgraphics/2020/03/16/coronavirus-maps/";

string html = await client.GetStringAsync("https://www.nytimes.com/interactive/2020/us/coronavirus-us-cases.html");
int start = html.IndexOf(ROOT_URI) + ROOT_URI.Length;
int end = html.IndexOf('/', start + 1);
string code = html.Substring(start, end - start);
display(code);
```

## Get the JSON Covid Data

```python
const string COUNTRY = "CAN"; // BRA, DEU, ESP, FRA, GBR, IND, ITA, MEX, USA
var json = await client.GetStringAsync($"{ROOT_URI}{code}/data/timeseries/en/{COUNTRY}.json");
display(json);
```
